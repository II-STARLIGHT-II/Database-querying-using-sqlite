Use `Pandas` to load the data available in the links above to dataframes. Use these dataframes to load data on to the database `FinalDB.db` as required tables.


```python
import pandas as pd
import sqlite3

# Establish SQLite connection
conn = sqlite3.connect('FinalDB.db')

# URLs for datasets
census_data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCensusData.csv"
schools_data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv"
crime_data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCrimeData.csv"

# Load datasets into DataFrames
census_df = pd.read_csv(census_data_url)
schools_df = pd.read_csv(schools_data_url)
crime_df = pd.read_csv(crime_data_url)

# Save DataFrames to SQLite tables
census_df.to_sql("CENSUS_DATA", conn, if_exists='replace', index=False)
schools_df.to_sql("CHICAGO_PUBLIC_SCHOOLS", conn, if_exists='replace', index=False)
crime_df.to_sql("CHICAGO_CRIME_DATA", conn, if_exists='replace', index=False)
```

## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
# Using sqlite3 to execute the query and fetch the result
cursor = conn.cursor()
cursor.execute("SELECT COUNT(*) AS total_crimes FROM CHICAGO_CRIME_DATA;")
result = cursor.fetchone()

# Display the total number of crimes
print("Total number of crimes recorded:", result[0])

```

### Problem 2

##### List community area names and numbers with per capita income less than 11000.


```python
# Query to find community areas with per capita income less than 11000
query = """
SELECT community_area_number, community_area_name
FROM CENSUS_DATA
WHERE per_capita_income < 11000;
"""

# Execute the query
cursor.execute(query)
results = cursor.fetchall()

# Display the results
for row in results:
    print("Community Area Number:", row[0], "| Community Area Name:", row[1])
```

### Problem 3

##### List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis) 


```python
# Query to find all case numbers for crimes involving minors
query = """
SELECT case_number
FROM CHICAGO_CRIME_DATA
WHERE description LIKE '%minor%' OR description LIKE '%juvenile%';
"""

# Execute the query
cursor.execute(query)
results = cursor.fetchall()

# Display the results
for row in results:
    print("Case Number:", row[0])
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
# Query to find all kidnapping crimes involving a child
query = """
SELECT case_number, primary_type, description
FROM CHICAGO_CRIME_DATA
WHERE primary_type = 'KIDNAPPING' AND description LIKE '%child%';
"""

# Execute the query
cursor.execute(query)
results = cursor.fetchall()

# Display the results
for row in results:
    print("Case Number:", row[0], "| Primary Type:", row[1], "| Description:", row[2])
```

### Problem 5

##### List the kind of crimes that were recorded at schools. (No repetitions)


```python
# Query to list distinct types of crimes recorded at schools
query = """
SELECT DISTINCT primary_type
FROM CHICAGO_CRIME_DATA
WHERE location_description LIKE '%school%';
"""

# Execute the query
cursor.execute(query)
results = cursor.fetchall()

# Display the results
for row in results:
    print("Crime Type:", row[0])
```

### Problem 6

##### List the type of schools along with the average safety score for each type.


```python
# Query to get column names from CHICAGO_PUBLIC_SCHOOLS table
query = "PRAGMA table_info(CHICAGO_PUBLIC_SCHOOLS);"

# Execute the query to see the table structure
cursor.execute(query)
columns = cursor.fetchall()

# Display column names
for column in columns:
    print("Column Name:", column[1])
    
# Updated query to find the type of schools along with their average safety score
query = """
SELECT "Elementary, Middle, or High School" AS school_type, AVG(SAFETY_SCORE) AS average_safety_score
FROM CHICAGO_PUBLIC_SCHOOLS
GROUP BY "Elementary, Middle, or High School";
"""

# Execute the query
cursor.execute(query)
results = cursor.fetchall()

# Display the results
for row in results:
    print("School Type:", row[0], "| Average Safety Score:", row[1])
    

```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
# Query to list 5 community areas with the highest percentage of households below the poverty line
query = """
SELECT community_area_name, percent_households_below_poverty
FROM CENSUS_DATA
ORDER BY percent_households_below_poverty DESC
LIMIT 5;
"""

# Execute the query
cursor.execute(query)
results = cursor.fetchall()

# Display the results
for row in results:
    print("Community Area:", row[0], "| % Households Below Poverty:", row[1])
```

### Problem 8

##### Which community area is most crime prone? Display the coumminty area number only.


```python
# Query to find the most crime-prone community area number
query = """
SELECT community_area_number
FROM CHICAGO_CRIME_DATA
GROUP BY community_area_number
ORDER BY COUNT(*) DESC
LIMIT 1;
"""

# Execute the query
cursor.execute(query)
result = cursor.fetchone()

# Display the most crime-prone community area number
print("Most Crime-Prone Community Area Number:", result[0])
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
# Query to find the name of the community area with the highest hardship index
query = """
SELECT community_area_name
FROM CENSUS_DATA
WHERE hardship_index = (SELECT MAX(hardship_index) FROM CENSUS_DATA);
"""

# Execute the query
cursor.execute(query)
result = cursor.fetchone()

# Display the community area with the highest hardship index
print("Community Area with Highest Hardship Index:", result[0])
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
# Query to find the community area name with the most number of crimes
query = """
SELECT community_area_name
FROM CENSUS_DATA
WHERE community_area_number = (
    SELECT community_area_number
    FROM CHICAGO_CRIME_DATA
    GROUP BY community_area_number
    ORDER BY COUNT(*) DESC
    LIMIT 1
);
"""

# Execute the query
cursor.execute(query)
result = cursor.fetchone()

# Display the community area with the most crimes
print("Community Area with Most Crimes:", result[0])
```
